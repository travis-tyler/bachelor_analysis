### QUESTION: On the "Bachelorette" and "Bachelor," how important is the "first impression?"
What percentage of winners recieved the "first impression" rose?

Data source: https://www.kaggle.com/fivethirtyeight/fivethirtyeight-bachelorette-dataset?select=bachelorette.csv

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
# Read in data using pandas
df = pd.DataFrame(pd.read_csv('../resources/bachelorette.csv'))
df
```

```python
# Split data by show
bachelorette_df = df.loc[df['SHOW']=='Bachelorette']
bachelor_df = df.loc[df['SHOW']=='Bachelor']

# Create lists of relevant seasons
# (where the 'first impression rose' (R1) is used)
bachelorette_r1_seasons = bachelorette_df.loc[bachelorette_df['ELIMINATION-1']=='R1']['SEASON'].tolist()
bachelor_r1_seasons = bachelor_df.loc[bachelor_df['ELIMINATION-1']=='R1']['SEASON'].tolist()

# Filter both DFs by R1 lists
bachelorette_df = bachelorette_df[bachelorette_df['SEASON'].isin(bachelorette_r1_seasons)]
bachelor_df = bachelor_df[bachelor_df['SEASON'].isin(bachelor_r1_seasons)]

# Combine DFs, remove NaNs and reset the index
r1_seasons_df = bachelorette_df.append(bachelor_df).fillna('0').reset_index()
r1_seasons_df
```

```python
# Create a list containing season winners
# (first contestant listed for each season)
winners = r1_seasons_df.groupby(['SHOW','SEASON']).first()['CONTESTANT'].tolist()
print(winners)
```

```python
# Create new columns for first impression and winners
r1_seasons_df['first_impression'] = ''
r1_seasons_df['winner'] = ''

# Iterate through the DF
for index, row in r1_seasons_df.iterrows():
    
    # Change 'first_impression' to 'y' if contestant got 'first impression rose'
    if r1_seasons_df.loc[index,'ELIMINATION-1'] == 'R1':
        r1_seasons_df.loc[index,'first_impression'] = 'y'
    else:
        r1_seasons_df.loc[index,'first_impression'] = 'n'
        
    # Change 'winner' to 'y' if contestant won the season
    if r1_seasons_df.loc[index,'CONTESTANT'] in winners:
        r1_seasons_df.loc[index,'winner'] = 'y'
    else:
        r1_seasons_df.loc[index,'winner'] = 'n'

# Create a smaller, more readable summary containing only relevant data
first_last = r1_seasons_df[['SHOW','SEASON','CONTESTANT','first_impression','winner']]
    
first_last
```

```python
# Create a new column to denote the category of win or loss
first_last['final_status'] = ''

# Pandas conditionals
first_last.loc[(first_last['first_impression']=='y')&(first_last['winner']=='y'),'final_status']='First impression - Won'
first_last.loc[(first_last['first_impression']=='n')&(first_last['winner']=='y'),'final_status']='No first impression - Won'
first_last.loc[(first_last['first_impression']=='y')&(first_last['winner']=='n'),'final_status']='First impression - Lost'
first_last.loc[(first_last['first_impression']=='n')&(first_last['winner']=='n'),'final_status']='Lost'

first_last
```

```python
# Drop all rows containing the 'lost' condition, reorder based on season, and reset index
first_last = first_last[first_last.final_status != 'Lost'].sort_values('SEASON').reset_index(drop=True)
first_last
```

```python
# Some seasons listed too many 'first impression' recipients
first_last.groupby(['SHOW','SEASON','first_impression']).size()
# Identified Season 17 of the 'Bachelor' and Season 9 of the 'Bachelorette' 
# as having an incorrect number of 'first impression' recipients'
```

```python
# Created lists to hold all contestants from both offending seasons
contestants_17 = first_last.loc[(first_last.SEASON=='17')&(first_last.SHOW=='Bachelor')]['CONTESTANT'].tolist()
# Joined the lists
contestants = contestants_17 + first_last.loc[(first_last.SEASON=='09')&(first_last.SHOW=='Bachelorette')]['CONTESTANT'].tolist()

# Manually entered winners and rose recipients into a list
keep_contestants = ['17_TIERRA_L','17_CATHERINE_G','09_BEN_S','09_CHRIS_S']

# Used list comprehension to remove the correct names
remove_contestants = [x for x in contestants if x not in keep_contestants]

remove_contestants
```

```python
# Removed conestants from DF
first_last = first_last[~first_last['CONTESTANT'].isin(remove_contestants)].reset_index(drop=True)
first_last.head()
```

```python
# Create chart to display win/loss category counts across all shows and seasons
first_last.groupby('final_status').size().sort_values(ascending=False)\
    .plot.bar(title='"First impression rose" winners and losers - Both shows', figsize=(8,6), color='orange')
plt.xticks(rotation=45)
```

```python
# Create chart to display win/loss category counts across all seasons
first_last.loc[first_last['SHOW']=='Bachelor'].groupby('final_status').size().sort_values(ascending=False)\
    .plot.bar(title='"First impression rose" winners and losers - Bachelor', figsize=(8,6))
plt.xticks(rotation=45)
```

```python
# Create chart to display win/loss category counts across all seasons
first_last.loc[first_last['SHOW']=='Bachelorette'].groupby('final_status').size().sort_values(ascending=False)\
    .plot.bar(title='"First impression rose" winners and losers - Bachelorette', figsize=(8,6), color='pink')
plt.xticks(rotation=45)
```

```python
# Variables to hold win percentages for both shows and overall
overall_odds = len(first_last.loc[(first_last.winner=='y')&(first_last.final_status=='First impression - Won')])/\
    len(first_last.loc[first_last.winner=='y'])


bachelorette_odds = len(first_last.loc[(first_last.winner=='y')&(first_last.final_status=='First impression - Won')&(first_last['SHOW']=='Bachelorette')])/\
    len(first_last.loc[(first_last.winner=='y')&(first_last['SHOW']=='Bachelorette')])

bachelor_odds = len(first_last.loc[(first_last.winner=='y')&(first_last.final_status=='First impression - Won')&(first_last['SHOW']=='Bachelor')])/\
    len(first_last.loc[(first_last.winner=='y')&(first_last['SHOW']=='Bachelor')])
```

```python
# Create new DF and convert 'SEASON' to int
over_time = first_last[['SEASON','final_status']]
over_time = over_time.astype({'SEASON': 'int32'})
# Change win category to 0 or 1 and group by season
over_time = over_time.replace({'First impression - Lost':0,'First impression - Won':1,'No first impression - Won':0})
over_time = over_time.groupby('SEASON').sum()
```

```python
# Simply show which seasons were won by first impression recipient
# to visualize grouping
over_time.plot.bar(legend=None, title='First impression wins (both shows)', figsize=(8,6), color='r')
```

```python
# Figure running count and show increase over time
over_time_running = over_time.cumsum()
over_time_running.plot.line(legend=None, title='First impression wins (both shows)', figsize=(8,6))
```

# Findings

```python
print(f"Men win the 'Bachelorette' {round(bachelorette_odds*100, 1)}% of the time if they recieve the 'first impression rose'")
print("------------------------------------------------------------------------------------------------------")
print(f"Women win the 'Bachelor' {round(bachelor_odds*100, 1)}% of the time if they recieve the 'first impression rose'")
print("------------------------------------------------------------------------------------------------------")
print(f"Overall, contestants win {round(overall_odds*100, 1)}% of the time if they recieve the 'first impression rose'")
```

# Conclusion

On the shows "Bachelorette" and "Bachelor," the "first impression rose" is a reasonable predictor of success, with over one-fourth of all contestants who recieve it winning. 

The "first impression" is a much more accurate predictor for the "Bachelorette," with 45.5% of contestants winning the season vs. only 13.3% from the "Bachelor."

This could be a result of differences between what men and women look for at the start of a relationship (i.e. Men seeking out the most physically attractive women, which can differ from actual compatibility vs. women seeking out personality traits from the beginning).
